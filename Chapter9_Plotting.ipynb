```python
from datetime import datetime
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
# spx.plot(ax=ax, style='k-')
data = pd.read_csv(r'H:/Python/Examples/examples_data_for_data_analysis/spx.csv', index_col=0, parse_dates=True)
spx = data['SPX']
spx.plot(ax=ax, style='k-')

crisis_data = [(datetime(2007, 10, 11), 'Peak of bull market'),\
(datetime(2008, 3, 12), 'Bear Stearns Fails'),\
(datetime(2008, 9, 15), 'Lehman Bankruptcy')]

for date, label in crisis_data:
        ax.annotate(label, xy=(date, spx.asof(date) + 75), xytext=(date, spx.asof(date) + 225),\
arrowprops=dict(facecolor='black', headwidth=4, width=2,headlength=4),\
horizontalalignment='left', verticalalignment='top')

# Zoom in on 2007-2010
ax.set_xlim(['1/1/2007', '1/1/2011'])     
ax.set_ylim([600, 1800])
ax.set_title('Important dates in the 2008-2009 financial crisis')
```

```python
# add a shape to a plot, you create the patch object shp and add it to a subplot by calling ax.add_patch(shp)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color = 'r', alpha = 0.5)  #
circ = plt.Circle((0.7, 0.2), 0.15, color = 'b', alpha = 0.5)   # 
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], color = 'k', alpha = 0.3)  #
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```

```python
tips = pd.read_csv(r'H:/Python/Examples/examples_data_for_data_analysis/tips.csv')
party_counts = pd.crosstab(tips['day'], tips['size'])
party_counts
```

```python
party_counts = party_counts.loc[:, 2:5]
type(party_counts)
party_counts
```

```python
# Normalize to sum to 1

party_pcts = party_counts.div(party_counts.sum(1), axis=0)
party_pcts
```

```python
party_pcts.plot.bar()
```

```python
# With data that requires aggregation or summarization before making a plot, 
# using the seaborn package can make things much simpler

import seaborn as sns

tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips.head()
```

```python
sns.barplot(x='tip_pct', y='day', data=tips, orient='h')
```

```python
sns.barplot(x='tip_pct', y='day', hue='time', data=tips, orient='h')
```

```python
# Histograms and Density Plots

#  a histogram of tip percentages of the total bill using the plot.hist method on the Series

tips['tip_pct'].plot.hist(bins=50)
```

```python
# density plots are also known as kernel density estimate (KDE) plots. 
# Using plot.kde makes a density plot using the conventional mixture-of-normals estimate

tips['tip_pct'].plot.density()
```

```python
comp1 = np.random.normal(0, 1, size=200)
comp2 = np.random.normal(30, 3, size=200)
values = pd.Series(np.concatenate([comp1, comp2]))
sns.distplot(values, bins=100, color='k')
```

```python
# Scatter or Point Plots

macro = pd.read_csv(r'H:/Python/Examples/examples_data_for_data_analysis/macrodata.csv')
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]
trans_data = np.log(data).diff().dropna()
trans_data[-5:]
```

```python
sns.regplot('m1', 'unemp', data=trans_data)
plt.title('Changes in log %s versus log %s' % ('m1', 'unemp'))
```
